```python
# Import the csv file into dataframe with the seperator \t or tab
import pandas as pd
marketing_df = pd.read_csv("/Users/Darshan/Desktop/Projects/Marketing_analysis_roi/data/marketing_campaign_raw.csv", sep="\t")
marketing_df.head(15)
```

```python
marketing_df.columns
```

### Removed the Columns which are not relevant to our goal ###

```python
marketing_df.drop(columns=['Z_CostContact', 'Z_Revenue', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5'], inplace=True)
```

```python
marketing_df.info()
```

### Convert date into datetime format ###

```python
marketing_df['Dt_Customer'] = pd.to_datetime(marketing_df['Dt_Customer'], format='%d-%m-%Y', dayfirst=True)
```

```python
marketing_df.info()
```

```python
marketing_df.head(15)
```

### Create column total spent to get the value of total spendings ###

```python
marketing_df['Total_Spent'] = marketing_df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)
```

```python
marketing_df.head(15)
```

### Modernize the data by changing the year to simulate more modern ###

```python
marketing_df['Dt_Customer'] = marketing_df['Dt_Customer'].apply(lambda x: x.replace(year=2023))
marketing_df['Tenure_Months'] = ((pd.Timestamp('2024-12-31') - marketing_df['Dt_Customer']).dt.days // 30).clip(lower=1)
```

```python
marketing_df.head(15)
```

### Simulate Ad Spend ###

```python
import numpy as np
np.random.seed(42)
marketing_df['Ad_Spend'] = np.random.normal(loc=50, scale=15, size=len(marketing_df)).clip(min=10)
```

### Estimate Revenue ###

```python
marketing_df['Revenue'] = marketing_df['Total_Spent'] * np.random.uniform(1.2, 1.6, size=len(marketing_df))
```

### Tag Churn ###

```python
marketing_df['Churn'] = (marketing_df['Recency'] > 80) & (marketing_df['Response'] == 0)
```

### Assign Channel ###

```python
marketing_df['Channel'] = np.random.choice(['Email', 'Phone', 'SMS', 'Social Media'], size=len(marketing_df), p=[0.3, 0.4, 0.2, 0.1])
```
